```python
# STEP 4: Create patient split (30 patients)

import os
import json

print("=== STEP 4: Creating patient split ===")

# 1. Get ALL patients from your images folder
all_patients = os.listdir('/content/input/images/images')
print(f"Found {len(all_patients)} total patients")

# 2. Select first 30 patients automatically
selected_30 = sorted(all_patients)[:30]
print(f"Selected 30 patients: {selected_30[:5]}...")

# 3. Create train/validation split: 20 train, 10 validation
split_config = {
    "fold_0": {
        "train": selected_30[:20],   # First 20 for training
        "val": selected_30[20:]      # Last 10 for validation
    }
}

# 4. Save this split to file
os.makedirs('/content/workdir/splits', exist_ok=True)
split_file = '/content/workdir/splits/custom.json'
with open(split_file, 'w') as f:
    json.dump(split_config, f, indent=2)

print(f"\n‚úÖ Split created:")
print(f"   Training patients: {len(split_config['fold_0']['train'])}")
print(f"   Validation patients: {len(split_config['fold_0']['val'])}")
print(f"   Saved to: {split_file}")

# 5. Verify the file
print("\nüìÑ Split file preview:")
!head -20 "$split_file"
```

```python
print("=== Verify ALL 30 patients ===")
import json

with open('/content/workdir/splits/custom.json') as f:
    data = json.load(f)

print(f"Total train patients: {len(data['fold_0']['train'])}")
print(f"Total val patients: {len(data['fold_0']['val'])}")

print("\nFirst 10 train patients:")
print(data['fold_0']['train'][:10])

print("\nAll validation patients (10):")
print(data['fold_0']['val'])
```
